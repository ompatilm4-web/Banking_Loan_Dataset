```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
Data = pd.read_csv('/home/omnix/Desktop/Banking_Loan/Banking_Loan_Dataset/Data/loan_default_raw_dataset.csv')
Data.head()
```

```python
Data.info()
```

```python
Data.isnull().sum()
```

```python
Data['dependents']=Data['dependents'].astype(int)
Data.drop('application_date',axis=1,inplace=True)
```

```python
Data.drop(Data[Data['dependents']=='3+'].index,inplace=True)
```

```python
Data['customer_id'] = Data['customer_id'].str.split('_').str[1]
Data['customer_id']=Data['customer_id'].astype(int)
```

```python
Data.isnull().sum()
```

```python
sns.histplot(Data['employment_years'],kde=True)
# Data.drop(Data[Data['employment_years']==0].index,inplace=True)
Data['employment_years'] = Data['employment_years'].fillna(Data['employment_years'].median())
```

```python
Data['annual_income'].dtype
sns.histplot(Data['annual_income'],kde=True)
plt.ticklabel_format(style='plain', axis='x')  # This is the key line

Data['annual_income'] = Data['annual_income'].fillna(Data['annual_income'].median())
```

```python
Data['monthly_debt'].dtype
sns.histplot(Data['monthly_debt'],kde=True)
plt.ticklabel_format(style='plain',axis='x')

Data['monthly_debt'] = Data ['monthly_debt'].fillna(Data['monthly_debt'].mean())
Data.isnull().sum()
```

```python
Data['credit_score'].dtype
sns.histplot(Data['credit_score'],kde=True)
plt.ticklabel_format(style='plain',axis='x')


Data['credit_score'] = Data['credit_score'].fillna(Data['credit_score'].mean())
Data.isnull().sum()
```

```python
Data.drop('debt_to_income_ratio',axis=1,inplace=True)
```

```python
Data.info()
```

```python
Data
```

```python
Data['Loan_issue_Year'] = Data['loan_issue_date'].str.split('-').str[0].astype(int)
```

```python
Data['Loan_issue_Month'] = Data['loan_issue_date'].str.split('-').str[1].astype(int
                                                                                )
```

```python
Data.drop('loan_issue_date',axis=1,inplace=True)
```

```python
Data.drop('loan_purpose',axis=1,inplace=True)
```

```python
Data.drop('education',axis=1,inplace=True)
```

```python
Data.drop('employment_years',axis=1,inplace=True)
```

```python
Data.drop('loan_to_income_ratio',inplace=True,axis=1)
```

```python
Data['loan_retained'] = Data['loan_amount'] * (Data['default'] == 1).astype(int)
Data['interest_retained'] = Data['loan_retained'] * (Data['interest_rate'] / 100) * (Data['loan_term_months'] / 12)
```

```python
Data
```

```python
Data['default']=Data['default'].map({1:'returned',0:'Not_Returned'})
```

```python
Data.drop('default',axis=1,inplace=True)
```
