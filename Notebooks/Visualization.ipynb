```python
# Create loan brackets
bins = [0, 10000, 25000, 50000, 75000, 100000, 500000]
labels = ['<10K', '10K-25K', '25K-50K', '50K-75K', '75K-100K', '100K+']
Data['loan_bracket'] = pd.cut(Data['loan_amount'], bins=bins, labels=labels)

# Aggregated analysis
loan_income_analysis = Data.groupby('loan_bracket').agg({
    'annual_income': ['mean', 'median', 'min', 'max', 'count'],
    'loan_amount': ['mean', 'count'],
    'interest_retained': 'mean'
}).round(2)

print(loan_income_analysis)

# Bar chart (with only 6 groups - readable!)
loan_income_analysis[('annual_income', 'mean')].plot(kind='bar', figsize=(10, 6))
plt.title('Average Annual Income by Loan Amount Bracket')
plt.xlabel('Loan Amount Range')
plt.ylabel('Average Annual Income ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# WHO generates the MOST interest?
profitability = Data.groupby('employment_status').agg({
    'interest_retained': ['sum', 'mean', 'count'],
    'loan_amount': 'sum'
}).round(2)

profitability.columns = ['Total_Interest', 'Avg_Interest', 'Customer_Count', 'Total_Loans']
profitability['Interest_per_Customer'] = profitability['Total_Interest'] / profitability['Customer_Count']
profitability = profitability.sort_values('Total_Interest', ascending=False)

print("="*60)
print("MOST PROFITABLE CUSTOMER SEGMENTS")
print("="*60)
print(profitability)
print("\n💡 INSIGHT: The segment with highest Total_Interest is your PRIMARY TARGET")
```

```python
# WHO is most risky?
risk = Data.groupby('employment_status').agg({
    'late_payments_90plus_days': 'sum',
    'num_collections': 'sum',
    'interest_retained': 'sum',
    'customer_id': 'count'
}).round(2)

risk.columns = ['Late_Payments', 'Collections', 'Total_Interest', 'Customer_Count']
risk['Risk_per_Customer'] = (risk['Late_Payments'] + risk['Collections']) / risk['Customer_Count']
risk['Interest_per_Risk'] = risk['Total_Interest'] / (risk['Late_Payments'] + risk['Collections'] + 1)
risk = risk.sort_values('Risk_per_Customer', ascending=False)

print("="*60)
print("RISK ASSESSMENT BY SEGMENT")
print("="*60)
print(risk)
print("\n💡 INSIGHT: High Risk_per_Customer = Need stricter lending criteria")
```

```python
# Create interest tiers
Data['interest_tier'] = pd.cut(Data['interest_retained'], 
                                bins=[-1, 0, 5000, 15000, 30000, 1000000],
                                labels=['Zero', 'Low', 'Medium', 'High', 'Very High'])

# Profile comparison
high_value = Data[Data['interest_tier'] == 'Very High'].agg({
    'annual_income': 'mean',
    'loan_amount': 'mean',
    'interest_rate': 'mean',
    'age': 'mean',
    'loan_term_Years': 'mean',
    'monthly_debt': 'mean'
}).round(2)

low_value = Data[Data['interest_tier'] == 'Low'].agg({
    'annual_income': 'mean',
    'loan_amount': 'mean',
    'interest_rate': 'mean',
    'age': 'mean',
    'loan_term_Years': 'mean',
    'monthly_debt': 'mean'
}).round(2)

comparison = pd.DataFrame({
    'High_Value_Customers': high_value,
    'Low_Value_Customers': low_value,
    'Difference': high_value - low_value,
    '%_Difference': ((high_value - low_value) / low_value * 100).round(1)
})

print("="*70)
print("WHAT MAKES A HIGH-VALUE CUSTOMER?")
print("="*70)
print(comparison)
```

```python
# Identify customers ready for loan upsell
upsell_candidates = Data[
    # Financial capacity
    (Data['annual_income'] > 100000) &  # High income
    (Data['loan_amount'] < 50000) &  # Small current loan
    (Data['monthly_debt'] < 4000) &  # Manageable debt
    
    # Low risk indicators
    (Data['late_payments_90plus_days'] < 3) &  # Good payment history
    (Data['num_collections'] < 3) &  # No collections
    
    # Employment stability
    (Data['employment_status'].isin(['Employed', 'Self-Employed']))
]

print("="*70)
print("UPSEL OPPORTUNITY: LOAN AMOUNT INCREASE")
print("="*70)
print(f"Total candidates: {len(upsell_candidates):,} customers")
print(f"Current total loans: ${upsell_candidates['loan_amount'].sum():,.0f}")
print(f"Potential increase: ${upsell_candidates['loan_amount'].sum() * 2.5:,.0f}")
print(f"Potential additional interest: ${upsell_candidates['loan_amount'].sum() * 0.15:,.0f}")
print("\nSample candidates:")
print(upsell_candidates[['customer_id', 'loan_amount', 'annual_income', 
                        'employment_status', 'interest_retained']].head(10))
```
